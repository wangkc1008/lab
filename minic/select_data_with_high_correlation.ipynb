```python
head_200, other_data = np.array(data)[:, :200], np.array(data)[:, 200:]
```

```python
other_data = pd.DataFrame(other_data)
other_data
```

```python
other_data.rename(columns=({108: 'flag'}), inplace=True)
other_data
```

```python
first_row = other_data.iloc[0, :]
first_row
```

```python
corr = other_data.corr()
flag_corr_sort = corr['flag'].sort_values(ascending=False)
column_corr = flag_corr_sort[((flag_corr_sort > 0.1) | (flag_corr_sort < -0.1))]
column_corr
```

```python
corr_data = other_data[column_corr.index]
corr_data
```

```python
corr_first_row = corr_data.iloc[0, :]
for index in corr_first_row.index:
    print(first_row[index], corr_first_row[index]) #主要是想看看还是不是这一行
```

```python
corr_array = np.array(corr_data)
corr_array.shape
```

```python
total_data = np.hstack((head_200, corr_array))
total_data.shape
```

```python
total_data = pd.DataFrame(total_data)
total_data.head()
```

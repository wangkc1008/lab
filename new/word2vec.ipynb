```python
res_list = []
res_text = text[]
for item in text:
    new_text = re.sub('\[[^\[\]]*\]', ' ', item)
    new_text = re.sub('[#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”！[\\]^_`{|}~\s]+', " ", new_text)
    new_text = new_text.replace('\r\n', ' ').replace('\n', ' ').replace('\r', ' ').split(' ')
    new_text = list(filter(None, new_text))
    res_list.append(new_text)
len(res_list)
```

```python
res_list[5000]
```

```python
start = time.clock()
model=word2vec.Word2Vec(res_list, size=200, sg=1)
end = time.clock()
print('Running time: %s Seconds'%(end-start))
```

```python
model.most_similar('pressure')
```

```python
model.save('./AMI_word.model')
```

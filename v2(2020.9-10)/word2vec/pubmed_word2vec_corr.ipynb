```python
import gensim
import pandas as pd
import numpy as np
import re
```

```python
embedding_path = 'E:/hxf_prediction/data_mimic/hxf/notes_extract/total_code/wikipedia-pubmed-and-PMC-w2v.bin'
word2vec = gensim.models.KeyedVectors.load_word2vec_format(embedding_path,binary=True,size=300)
```

```python
print('Found %s word vectors of word2vec' % len(word2vec.vocab))
```

```python
file_name = 'E:/hxf_prediction/data_mimic/hxf/notes_extract/total_code/text_flag_20200714.csv'
data = pd.read_csv(file_name)
```

```python
text = data[['hadm_id', 'text']].values
text[0, 0]
print(text)
```

```python
res_list = []
for hadm_id, item in text:
    new_text = re.sub('\[[^\[\]]*\]', ' ', item)
    new_text = re.sub('[#$%&\'()*+,-./:;<=>?@?[\\]^_`{|}~\s]+', " ", new_text)
    new_text = new_text.replace('\r\n', ' ').replace('\n', ' ').replace('\r', ' ').split(' ')
    new_text = list(filter(None, new_text))
    res_list.append([hadm_id, new_text])
len(res_list)
```

```python
vocab_data = []
for item in text: 
    res = np.zeros(200)
    num = 0
    none_num = 0
    for word in item[1]:
        if word in word2vec.wv.vocab:
            res += word2vec[word]
            num += 1
        elif word.lower() in word2vec.wv.vocab:
            print('lower: ', item[0], word)
            res += word2vec[word.lower()]
            num += 1
        elif word.upper() in word2vec.wv.vocab:
            print('upper: ', item[0], word)
            res += word2vec[word.upper()]
            num += 1
        else:
#             print(item[0], word)
            none_num += 1
    print('', item[0], num)
    print('None', item[0], none_num)
    vocab_data.append(np.append(item[0], res / num))
vocab_data = np.array(vocab_data)
```

```python
len(vocab_data)
```

```python
vocab_pd = pd.DataFrame(vocab_data)
vocab_pd
```

```python
vocab_pd['hadm_id'] = [int(item) for item in vocab_pd[0].values]
vocab_pd = vocab_pd.drop(labels=[0], axis=1)
```

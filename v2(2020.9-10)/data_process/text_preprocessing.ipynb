```python
text = data[['hadm_id', 'text']].values
text.shape
```

```python
hadm_id_list = []
for hadm_id, content in text:
    res = re.search('ECHO\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res is None:
        hadm_id_list.append(hadm_id)
print(len(hadm_id_list))
print(hadm_id_list)
```

```python
for hadm_id, content in text:
    res = re.search('ECHO\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res is not None:
        print(hadm_id, res.group())
```

```python
for hadm_id, content in text:
    res = re.search('CXR\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res is not None:
        print(hadm_id, res.group())
```

```python
# text[0, 1]
```

```python
res_list = {}
for hadm_id, content in text:
    res1 = re.search('Major Surgical or Invasive Procedure:[\s\S]*?History of Present Illness:', content, re.IGNORECASE)
    if res1 is not None:
#         print(hadm_id, res.group().split('History of Present Illness:')[0])
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res1.group().split('History of Present Illness:')[0]
    
    res2 = re.search('History of [Present Illness:|the Present Illness:][\s\S]*?Past Medical History', content, re.IGNORECASE)
    if res2 is not None:
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res2.group()
        else:
            res_list[hadm_id] = res_list[hadm_id] + ' ' + res2.group()
    
    res3 = re.search('Past Medical History[\s\S]*?\n\n', content, re.IGNORECASE)
    if res3 is not None:
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res3.group()
        else:
            res_list[hadm_id] = res_list[hadm_id] + ' ' + res3.group()
    
    res4 = re.search('ECG\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res4 is not None:
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res4.group()
        else:
            res_list[hadm_id] = res_list[hadm_id] + ' ' + res4.group()
            
    res5 = re.search('CXR\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res5 is not None:
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res5.group()
        else:
            res_list[hadm_id] = res_list[hadm_id] + ' ' + res5.group()
    
    res6 = re.search('ECHO\s\[\*[\s\S]*?\n\n', content, re.IGNORECASE)
    if res6 is not None:
        if hadm_id not in res_list.keys():
            res_list[hadm_id] = res6.group()
        else:
            res_list[hadm_id] = res_list[hadm_id] + ' ' + res6.group()

print(len(res_list))
```
